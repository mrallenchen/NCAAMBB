```python
import pickle
gamelog = pickle.load(open('gamelog.pickle', "rb" ))
advanced_gamelog = pickle.load(open('advanced_gamelog.pickle', "rb" ))
# Import Counter module
from collections import Counter
```

```python
 # Merge the values of dictionaries based on common keys
combo_log = {}
for game in gamelog.keys():
    gamelog[game].extend(advanced_gamelog[game])
```
